```python
# Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year.
# You have to write a program to find out the population at the end of each of the last 10 years.
currentpop=10000
for i in range(10,0,-1):
    currentpop=currentpop-(0.1*currentpop)
    print("year:", i , "population:", currentpop)

```

```python
# Sequence sum
# 1/1! + 2/2! + 3/3! + ...

n=int(input("enter a number"))
sum=0
fact=1
for i in range(1,n+1):
    fact=fact*i
    sum+=i/fact
print(sum)

```

**NESTED LOOPS**

```python
# unique pairs
for i in range (1,11):
    for j in range(1,11):
        print(i,j)
```

```python
# Pattern 2
# *
# **
# ***

for i in range(1,4):
    for j in range(1,i+1):
        print("*", end='') #prints in the same line
    print() # helps not printing everything in the same line
```

```python
#Pattern 2
#1
#121
#12321
#1234321

for i in range(0,5):
    for j in range(1,i+1):
        print(j, end='')
    for k in range(i-1,0,-1):
        print(k, end='')
    print()

```

**LOOP CONTROL STATEMENTS**
- break
- continue
- pass

```python
# break 
# if condition meets, doesnt print that and goes back to starting of the loop after printing the thing below it. continue doesnt, goes strsight back to start of loop
# real life example: searching for a person in database. once u found that person u can stop. useless going further. so break needed there

for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
# print only prime numbers 

lowerrange= int(input("enter the lower range"))
upperrange=int(input("enter upper range"))
for i in range(lowerrange, upperrange+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
```

```python
# continue
for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
# pass - used to store function or something for later use. says dont throw error rn, ill write the full code for it later

for i in range(1,11):
    pass
```

```python
# Count the frequency of a particular character in a provided string. 
# Eg 'hello how are you' is the string, the frequency of h in this string is 2.
string1=input('enter the string')
char1=input('enter the character to be searched')
count=0
for i in string1:
    if i==char1:
        count=count+1
print(count)
```

```python
# Write a program that can check whether a given string is palindrome or not.
string1=input('enter the string')
strlen=len(string1)
if string1[0:strlen]==string1[::-1]:
    print('it is a palindrome')
else:
    print('it is not a palindrome')
```
