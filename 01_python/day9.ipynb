### Exceptions

If things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened.

- Exceptions are raised by python runtime
- You have to takle is on the fly

#### **Examples**

- Memory overflow
- Divide by 0 -> logical error
- Database error

why exception handling is necessary
- doesnt scare the user seeing big technical errors thrown at them
- doesnt help a hacker with the code info leading to a breach

```python
# how to handle exceptions
# -> Try except block
```

```python
# let's create a file
with open('sample.txt','w') as f:
  f.write('hello world')
  # now what if user tries to read this file but names it wrong, says sample9 instead? it will give FileNotFoundError
```

```python
# try catch demo
try:
    with open('sample9.txt','r') as f:
        print(f.read())

except:
    print('sorry file not found')
```

```python
# catching specific exception
try:
 # m=5
  f = open('sample1.txt','r')
  print(f.read())
  print(o)
  print(5/0)
  L = [1,2,3]
  L[100]

except FileNotFoundError:
  print(' cant find the file')
except NameError:
  print('variable not defined')
except ZeroDivisionError:
  print("can't divide by 0")

except Exception as e:
  print(e)
```

```python
# Python exits the try block at the FIRST exception.
print(f.read())   # ✅ runs fine
print(o)          # ❌ NameError raised HERE → jumps out immediately
print(5/0)        # ❌ never reached
L = [1,2,3]       # ❌ never reached
L[100]            # ❌ never reached
```

```python
# better way
try:
    print(o)
except NameError:
    print('variable not defined')

try:
    print(5/0)
except ZeroDivisionError:
    print("can't divide by 0")
```

### Else
- done when u want to specify what to do if the code shows no errors

```python
try:
    f=open('sample.txt','r')
except FileNotFoundError:
    print('file nahi mili')
except Exception:
    print('some error')
else:
    print(f.read())
```

### finally
- done when u want something to happen irrespective of whether theres error or not
- usually used when file is to be closed or anything else 

```python
try:
    f=open('sample.txt','a')
except FileNotFoundError:
    print('file nahi mili')
except Exception:
    print('some error')
else:
    f.write('um ok')
finally:
    f.close()

#f.write('ok bye') wouldve given error if finally wasnt written
```

### Raise
raise lets you manually trigger an exception yourself, instead of waiting for Python to raise one naturally.


```python
class Bank:
    def __init__(self,balance):
        self.balance=balance

    def withdraw(self,amount):
        if amount<0:
            raise Exception ('negative not allowed')
        if self.balance<amount:
            raise Exception ('paise nahi hai tere paas')
        self.balance=self.balance-amount
        
obj=Bank(1000)
try:
    obj.withdraw(15000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
```

Creating a custom exception - your own exception

```python
class MyException(Exception):
    def __init__(self,message):
        print(message)
class Bank:

  def __init__(self,balance):
    self.balance = balance

  def withdraw(self,amount):
    if amount < 0:
      raise MyException('amount cannot be -ve')
    if self.balance < amount:
      raise MyException('paise nai hai tere paas')
    self.balance = self.balance - amount

obj=Bank(10000)
try:
   obj.withdraw(15000)
except MyException as e:
   pass
else:
   print(obj.balance)
```

```python
# why custom exception classes? 
# because you would want some other functionality to come with your custom exception class as well other than just printing the error msg.
# as in the one below youre adding the logout functionality upon error, not just a single messge
# thats why custom messages are usually thrown
```

```python
class SecurityError(Exception):
    def __init__(self,message):
        print(message)

    def logout(self):
        print('logout')

class Google:

  def __init__(self,name,email,password,device):
    self.name = name
    self.email = email
    self.password = password
    self.device = device

  def login(self,email,password,device):
    if device != self.device:
      raise SecurityError('bhai teri to lag gayi')
    if email == self.email and password == self.password:
      print('welcome',end=' ')
    else:
      print('login error')



obj = Google('nitish','nitish@gmail.com','1234','android')

try:
  obj.login('nitish@gmail.com','1234','android')
except SecurityError as e:
   e.logout()
else:
  print(obj.name)
finally:
  print('database connection closed')
```
