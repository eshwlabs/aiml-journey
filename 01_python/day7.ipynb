# Parameters Vs Arguments

Types of Arguments
- Default Argument
- Positional Argument

```python
# default argument - in case someone forgets to put value in arguments so parameters already have a default value to use to avoid errors

def power(a=1,b=1):
    """This function returns the exponential result of 2 numbers given. first number would be raised to the second number
    created- 14/05/26
    """
    return a**b

power()
```

```python
# keyword argument - if u forget which place which parameter exists, use this

power(b=4,a=5)
```

# *args and **kwargs

```python
# *args - if you dont want to set a limit on the no of arguments and user can put as many as they want

def multiply(*args): # can put anything after * like *eshel, *whatever etc
    product=1
    for i in args:
        product=product*i
    return product


multiply(1,2,3,4,5,6)
```

```python
# **kwargs
# **kwargs allows us to pass any number of keyword arguments.
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary.
```

```python
def display(**kwargs):
    for (key, value) in kwargs.items():
        print(key, '->', value)

display(india='delhi', pakistan='islamabad',france='paris')
```

Points to remember while using *args and **kwargs
- order of the arguments matter(normal -> *args -> **kwargs) (SEE BELOW CODE)
- The words “args” and “kwargs” are only a convention, you can use any name of your choice 

```python
def display(name, *args, **kwargs):
    print('name:', name)
    print('args:', args)
    print('kwargs:',kwargs)

display("Eshel", "python", "ML", age=20, city="delhi")
```

# Lambda Function
- very imp for data science

### Lambda Function

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

image.png

```python
# x -> x^2

lambda x:x**2
```

```python
# x,y -> x+y

a=lambda x,y:x+y
print(a(5,2))
```

#### Diff between lambda vs Normal Function

- No name
- lambda has no return value(infact,returns a function)
- lambda is written in 1 line
- not reusable

Then why use lambda functions?<br>
**They are used with HOF**


```python
# check if a string has 'a'
a = lambda s:'a' in s
a('hello')
```

```python
# odd or even
a= lambda num: 'even' if num%2==0 else 'odd'
a(2)
```

### Higher order functions
- a function that recieves another function as parameter
- OR a function that returns a function

```python
def transform(f, L):
    output=[]
    for i in L:
        output.append(f(i))
    print(output)
    
L=[1,2,3,4]
transform(lambda a:a**2, L)
```

# MAP
- creates a new list based on the given list using the logic given by lambda 

```python
# square the items of a list
list(map(lambda x:x**2, [1,2,3,4]))
```

```python
# odd/even labelling of list items

L=[1,2,3,4]
list(map(lambda num: 'even ' if num%2==0 else 'odd', L))
```

```python
# fetch names from a list of dict

users = [
    {
        'name':'Rahul',
        'age':45,
        'gender':'male'
    },
    {
        'name':'Nitish',
        'age':33,
        'gender':'male'
    },
    {
        'name':'Ankita',
        'age':50,
        'gender':'female'
    }
]

list(map(lambda x:x['name'], users))
```

```python
# numbers greater than 5

L = [3,4,5,6,7]

list(map(lambda x:x>5,L)) # but we dont want it applied and shown to all elements in the list. we want it to filter and show only numbers who are 
# greater than 5. for that we have filter below
```

### Filter

```python
# numbers greater than 5

L = [3,4,5,6,7]
list(filter(lambda x:x>5,L))
```

```python
# fetch fruits starting with 'a'
fruits = ['apple','guava','cherry']
list(filter(lambda x:x.startswith('a'), fruits))
```

### Reduce
- performs operations on two numbers
- reduce needs 2 parameters because it's always combining the running result with the next item.
- reduce already works directly on a list — no need to wrap it in list() like you do with map and filter.

```python
# sum of all item
import functools

functools.reduce(lambda x,y:x+y, [1,2,3,4,5])
```

```python
# find min
from functools import reduce
reduce(lambda x,y: x if x<y else y, [23,11,45,10,1])
```
