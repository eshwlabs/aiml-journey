# **TUPLES**
- same as lists but immutable (cant be changed)

Creating tuples 

```python
t=(2)
print(type(t)) # -> gives int not tuple so wrong way of creating tuple. put comma in the end

t1=(2,)
print(type(t1))
```

```python
# homo tuple
t=(1,2,3,4)
print(t)

# hetero tuple
t2=(1,2,'hello',[1,2,3,4])
print(t2)

# using type conversion
t3='hello'
print(tuple(t3))
```

Accessing Items
- Indexing
- Slicing

```python
t3=(1,2,3,4)
print(t3)
print(t3[0])
print(t3[-1])

print(t3[1::])
```

Editing items 
- not possible

```python
t3[0]=100
# not possible -> gives error
```

Adding items 
- not possible

Deleting
- deleting entire tuple possible

```python
print(t3)
del t3

del t3[0] # -> doesnt work
```

Operations on tuples 
-  +/* --> t1+t2  /  t1*3
- membership --> 'a' in t3
- iteration --> for i in loop

- ALL THE SAME AS LIST

Functions
- len / min / max / sum / sorted -> all same as list

```python
tup=(1,2,3,4)

print(sum(tup))
```

Methods 
- count / index -> same as list

tuple unpacking -> assigning variables values in a special way

```python
a,b,c=(1,2,3)
print(a,b,c)
```

```python
a,b=(1,2,3)
print(a,b)  # -> GIVES ERROR SO LOOK BELOW
```

```python
# using *others 
a,b, *others=(1,2,3,4)
print(a,b)
print(others)
```

```python
# swapping
a=1
b=2
a,b=b,a
print(a)
print(b)
```

```python
# zipping tuples 
t1=(1,2,3,4)
t2=(5,6,7,8)

tuple(zip(t1,t2))
```

# **SETS**
- unordered
- mutable
- no duplicates
- cant contain mutable datatypes

**Creating sets**

```python
# empty

s={} # WRONGGG...THIS CREATES DICTIONARY NOT A SET

s=set() # RIGHTTTT - USE TYPE CONVERSION FOR CREATING SETS
```

```python
# 1d 
s={1,2,3,4}

# 2d
s={1,2,3,{4,5}} # WRONGGG CANT CONTAIN MUTABLE DATATYPES INSIDE. WILL GIVE ERRORRR
```

```python
# homo 
s={1,2,3,4}

# hetero
s1={1,'hello',4.5,(3,4)}
print(s1)

s2={1,'hello',4.5,True} # True wont be written in output as True=1 and 1 is already in the set and sets dont allow duplicate values
print(s2)

s3={111,222,333}
print(s3)
```

```python
# accessing items -> error as sets are unordered, dont follow indexing
s={1,2,3,4}
s[0] # -> gives error
```

```python
# editing items -> not possible (as indexing doesnt work )
s[0]=100
```

Adding items

```python
# add
s={1,2,3,4}
s.add(5)

print(s)
```

```python
# update -> helps adding multiple items

s={1,2,3,4}
s.update([5,6]) # put in list

print(s)
```

Deleting items

```python
# del

s={1,2,3,4}
del s # deletes the tuple

# del s[0] not possible
```

```python
# discard

s={1,2,3,4}
s.discard(4)
print(s)

s.discard(50) # doesnt give error if it exceeds, just rturns the same tuple
print(s)

```

```python
# remove

s={1,2,3,4}
s.remove(4)
print(s)

# if s.remove(50) -> gives error
```

```python
# pop -> deletes random element, cant specify | returns item that it deletes if not printed
s={1,2,3,4}
s.pop()
print(s)
```

```python
# clear -> deletes all items in set but keeps the set 
s={1,2,3,4}
s.clear()
print(s)
```

# set operations

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}

# union
print(s1 | s2) #prints all values of both sets only once even if they are repeating

# intersection
print(s1 & s2) #prints only common/repeating in both values

# difference
print(s2-s1) # removes values of s1 from s1

# symmetric difference 
print(s1 ^ s2) # prints everything except for common values
```

```python
# membership and loop for iteration same as lists and tuples
print(1 in s1)

for i in s1:
    print(i)
```

SET FUNCTIONS

```python
# len / sum / min / max / sorted -> all same as list
```

```python
# union with union keyword
s1={1,2,3,4,5}
s2={4,5,6,7,8}

print(s1.union(s2))
print(s1)
```

```python
# update -> does union but puts answer in s1
s1.update(s2)
print(s1)
print(s2)
```

```python
# same for intersection 
```

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}

print(s1.intersection(s2))
print()

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference / difference_update

s1={1,2,3,4,5}
s2={4,5,6,7,8}

print(s1.difference(s2))
print()

s1.difference_update(s2)
print()
```
