# **TUPLES**
- same as lists but immutable (cant be changed)

Creating tuples 

```python
t=(2)
print(type(t)) # -> gives int not tuple so wrong way of creating tuple. put comma in the end

t1=(2,)
print(type(t1))
```

```python
# homo tuple
t=(1,2,3,4)
print(t)

# hetero tuple
t2=(1,2,'hello',[1,2,3,4])
print(t2)

# using type conversion
t3='hello'
print(tuple(t3))
```

Accessing Items
- Indexing
- Slicing

```python
t3=(1,2,3,4)
print(t3)
print(t3[0])
print(t3[-1])

print(t3[1::])
```

Editing items 
- not possible

```python
t3[0]=100
# not possible -> gives error
```

Adding items 
- not possible

Deleting
- deleting entire tuple possible

```python
print(t3)
del t3

del t3[0] # -> doesnt work
```

Operations on tuples 
-  +/* --> t1+t2  /  t1*3
- membership --> 'a' in t3
- iteration --> for i in loop

- ALL THE SAME AS LIST

Functions
- len / min / max / sum / sorted -> all same as list

```python
tup=(1,2,3,4)

print(sum(tup))
```

Methods 
- count / index -> same as list

tuple unpacking -> assigning variables values in a special way

```python
a,b,c=(1,2,3)
print(a,b,c)
```

```python
a,b=(1,2,3)
print(a,b)  # -> GIVES ERROR SO LOOK BELOW
```

```python
# using *others 
a,b, *others=(1,2,3,4)
print(a,b)
print(others)
```

```python
# swapping
a=1
b=2
a,b=b,a
print(a)
print(b)
```

```python
# zipping tuples 
t1=(1,2,3,4)
t2=(5,6,7,8)

tuple(zip(t1,t2))
```

# **SETS**
- unordered
- mutable
- no duplicates
- cant contain mutable datatypes

**Creating sets**

```python
# empty

s={} # WRONGGG...THIS CREATES DICTIONARY NOT A SET

s=set() # RIGHTTTT - USE TYPE CONVERSION FOR CREATING SETS
```

```python
# 1d 
s={1,2,3,4}

# 2d
s={1,2,3,{4,5}} # WRONGGG CANT CONTAIN MUTABLE DATATYPES INSIDE. WILL GIVE ERRORRR
```

```python
# homo 
s={1,2,3,4}

# hetero
s1={1,'hello',4.5,(3,4)}
print(s1)

s2={1,'hello',4.5,True} # True wont be written in output as True=1 and 1 is already in the set and sets dont allow duplicate values
print(s2)

s3={111,222,333}
print(s3)
```

```python
# accessing items -> error as sets are unordered, dont follow indexing
s={1,2,3,4}
s[0] # -> gives error
```

```python
# editing items -> not possible (as indexing doesnt work )
s[0]=100
```

Adding items

```python
# add
s={1,2,3,4}
s.add(5)

print(s)
```

```python
# update -> helps adding multiple items

s={1,2,3,4}
s.update([5,6]) # put in list

print(s)
```

Deleting items

```python
# del

s={1,2,3,4}
del s # deletes the tuple

# del s[0] not possible
```

```python
# discard

s={1,2,3,4}
s.discard(4)
print(s)

s.discard(50) # doesnt give error if it exceeds, just rturns the same tuple
print(s)

```

```python
# remove

s={1,2,3,4}
s.remove(4)
print(s)

# if s.remove(50) -> gives error
```

```python
# pop -> deletes random element, cant specify | returns item that it deletes if not printed
s={1,2,3,4}
s.pop()
print(s)
```

```python
# clear -> deletes all items in set but keeps the set 
s={1,2,3,4}
s.clear()
print(s)
```

# set operations

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}

# union
print(s1 | s2) #prints all values of both sets only once even if they are repeating

# intersection
print(s1 & s2) #prints only common/repeating in both values

# difference
print(s2-s1) # removes values of s1 from s1

# symmetric difference 
print(s1 ^ s2) # prints everything except for common values
```

```python
# membership and loop for iteration same as lists and tuples
print(1 in s1)

for i in s1:
    print(i)
```

SET FUNCTIONS

```python
# len / sum / min / max / sorted -> all same as list
```

```python
# union with union keyword
s1={1,2,3,4,5}
s2={4,5,6,7,8}

print(s1.union(s2))
print(s1)
```

```python
# update -> does union but puts answer in s1
s1.update(s2)
print(s1)
print(s2)
```

```python
# same for intersection 
```

```python
s1={1,2,3,4,5}
s2={4,5,6,7,8}

print(s1.intersection(s2))
print()

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference / difference_update

s1={1,2,3,4,5}
s2={4,5,6,7,8}

print(s1.difference(s2))
print()

s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference/symmetric_difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s1.symmetric_difference(s2))
print()

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint / issubset / issuperset

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s1.isdisjoint(s2))
```

```python
# issubset / issuperset

s1 = {1,2,3,4,5}
s2 = {3,4,5}

print(s2.issubset(s1))

print(s1.issuperset(s2))
```

```python
# copy
s1={1,2,3,4}
s2=s1.copy()

print(s1)
print(s2)
```

Frozen sets
- immutable version of sets 

```python
# creating frozen sets

fs1= frozenset([1,2,3])
fs2=frozenset([4,5,6])

print(fs1|fs2)
```

```python
# what works and what does not
# works -> all read functions -> intersection, union etc can be done
# does't work -> write operations -> remove, pop etc doesnt work
# when to use - when u dont want user to make changes in the set
```

```python
# 2d frozen set - immutable inside allowed

fs= frozenset([frozenset([1,2,3])])
print(fs)
```

Set comprehension

```python
# print no from 3 to 10

{i for i in range(3,11)}
```

# **DICTIONARIES**
- collection of key value pairs

Characterstics:
- Mutable
- Indexing has no meaning
- keys can't be duplicated
- keys can't be mutable items, but values can be

Creating dictionaries

```python
# empty
d={}

# 1d
dict1={'name':'eshel', 'city':'delhi'}

#2d - dict of a student with details and subject having another dict with diff subjects and their marks as values

dict2={
    'name': 'eshel',
    'college':'bits',
    'sem':4,
    'subject':{
        'dsa':40,
        'english':50,
        'dbms':30
    }
}
print(dict2)
```

```python
# using sequence and dict fuction

dict1=dict([('name','eshel'),('college','bits')])
print(dict1)
```

```python
# duplicate keys
d5 = {'name':'nitish','name':'rahul'} # only updated value gets printed if duplicate
d5
```

```python
# mutable items as keys not allowed, immutable allowed 
d6 = {'name':'nitish',(1,2,3):2} # tuples are immutable so allowed
print(d6)
```

Accessing items

```python
my_dict = {'name': 'Jack', 'age': 26}

# []
my_dict['age']

# get
my_dict.get('age')

# indexing not allowed
# my_dict[0] XXXXXXXXXXXXX
```

```python
# to access 2d dict
dict2={
    'name': 'eshel',
    'college':'bits',
    'sem':4,
    'subject':{
        'dsa':40,
        'english':50,
        'dbms':30
    }
}

print(dict2['subject']['english'])
```

Adding key value pair

```python
# adding key value pair in the dict
dict1 = {'name': 'Jack', 'age': 26}
dict1['college']='bits'
print(dict1)
```

Removing key value pair

```python
# pop
dict1 = {'name': 'Jack', 'age': 26}
dict1.pop('name')
print(dict1)
```

```python
# popitem - deletes last key value pair
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
dict1.popitem()
print(dict1)
```

```python
# del - deletes full dict or specific key value pair
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
del dict1['name']
print(dict1)
```

```python
# to delete from 2d dict
dict2={
    'name': 'eshel',
    'college':'bits',
    'sem':4,
    'subject':{
        'dsa':40,
        'english':50,
        'dbms':30
    }
}

del dict2['subject']['english']
print(dict2)
```

```python
# clear 
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
dict1.clear()
print(dict1)
```

Editing key value pairs

```python
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
dict1['name']='eshel'
print(dict1)
```

operations

```python
# membership 
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
print('Jack' in dict1) # values cant be put, only keys

print('name' in dict1)
```

```python
# iteration 
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
for i in dict1:
    print(i, dict1[i])
```

Dictionary functions

```python
# len
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
print(len(dict1))
```

```python
# sorted
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
#dict1.sorted() wrong syntax
print(sorted(dict1))
print(dict1)
```

```python
# items / keys / values
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}

print(dict1.items())
print(dict1.keys())
print(dict1.values())
```

```python
# update - updating a given dictionary using another dictionary
dict1={'name': 'Jack', 'age': 26, 'college': 'bits'}
dict2={'name':'eshel','college':'iit'}

dict1.update(dict2)
print(dict1)
```

dict comprehension

```python
# print 1st 10 numbers and their squares
{i:i**2 for i in range(1,11)}
```

```python
distances = {'delhi':1000,'mumbai':2000,'bangalore':3000}
print(distances.items())
# gives pairs in tuple - needed for below problem
```

```python
# covert distances to kms
distances = {'delhi':1000,'mumbai':2000,'bangalore':3000}

{key:value*0.62 for (key,value) in distances.items()}
```

```python
# using zip
days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

d1={key:value for (key,value) in zip(days, temp_C)}
print(d1)
```

```python
# using if condition
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

{key:value for (key,value) in products.items() if value>0}
```

```python
# Nested Comprehension
# print tables of number from 2 to 4

#{i*j for i in range(2,5) for j in range(1,11)}

{i:{j:i*j for j in range(1,11)} for i in range(2,5)} # og loop from down to up
```
