# **LISTS**
- used to store things of the same type in it efficiently 
- dynamically sized unlike java(fixed size)  where u have to specify the size beforehand eg -> int arr[50]
- order matters here. if u compare list1=[1,2,3] and list2=[3,2,1], it wont be equal as order isnt same.

```python
# lists vs arrays
# flexible sized vs fixed sized
# heterogenous container vs homogenous container
# but lists take more time
```

```python
# Characterstics of a List
# Ordered
# Changeble/Mutable
# Hetrogeneous
# Can have duplicates
# are dynamic
# can be nested
# items can be accessed
# can contain any kind of objects in python 
```

Creating a list

```python
# empty list
print([])
# 1D list
print([1,2,3,4,5])
# 2D list
print([1,2,3,[1,2,3]])
# 3d list
print([[[1,2],[3,4],[5,6]]]) #to count dimension of list, cound starting/ending brackets 
# heterogeneous list
print([1, True, 'hello'])
# type conversion -> converts any datatype stuff to list
print(list('hello'))
```

Accessing items from a list

```python
# indexing
L=[1,2,3,[4,5]]
print(L[3][0]) 

L=[[[1,2],[3,4]],[[5,6],[7,8]]]
print(L[1][0][0]) # no of boxes here should be equal to no of dimensions in the list 
```

```python
# slicing -> extracting multiple items at once
L=[1,2,3,4,5,6]
print(L[-6:-1])
```

**LIST METHODS**
- mutates og list

Adding items to a list

```python
# append -> adds one item to the end of the list
l=[1,2,3,4,5,6]
l.append(7)
print(l)
```

```python
# extend -> adds multiple items to the end 
l.extend([8,9,10])
print(l)

list2=[1,2,3,4,5]
list2.extend('delhi')
print(list2)
```

```python
# insert -> adds element at any index
l.insert(2,100)
print(l)
```

Editing lists

```python
l=[1,2,3,4,5,6] 

# editing using indexing :
l[2]=300
print(l)

# editing using slicing :
l[3:5]=[400,500]
print(l)
```

Deleting items from a list

```python
# del 

l=[1,2,3,4,5,6] 

# through indexing
del l[-1]
print(l)

# through slicing 
del l[1:3]
print(l)

# or del l to delete the entire list
```

```python
# remove -> removes an element without having to specify index position

l=[1,2,3,4,5,6] 
l.remove(6)
print(l)

#l.remove(4,5) -> error. only one element allowed
```

```python
# pop -> deletes using index, if no index-> deletes last item
l.pop()
print(l)
```

```python
# clear -> makes the list empty. only deletes the elements in it, not the list
l.clear()
print(l)
```

Operations on lists 
- arithmetic (+,*)
- membership

```python
# arithmetic 
l1=[1,2,3]
l2=[4,5,6]

print(l1+l2)
print(l1*3)
```

```python
# membership
l1=[1,2,3,4]
l2=[1,2,3,4,[5,6]]

print(4 in l1)
print(5 in l2)
print([5,6] in l2)
```

```python
# loops
l1=[1,2,3,4]
l2=[1,2,3,4,[5,6]]
l3=[[[1,2],[3,4]],[[5,6],[7,8]]]

for i in l1:
    print(i)
print()

for i in l2:
    print(i)
print()

for i in l3:
    print(i)
print()
```

**LIST FUNCTIONS**
- mutates copy list, not og list

```python
# len / min / max / sorted / sum

l1=[1,2,3,4,5,6]

print(len(l1))

print(min(l1))

print(max(l1))

print(sorted(l1)) # changes the copy of the list, not the og list

print(sorted(l1,reverse=True)) 

print(sum(l1))
```

```python
# count -> counts no of item appearing in that list
l1=[1,2,3,4,5,6]
l1.count(3)
```

```python
# index -> returns index of a specified item
l1=[1,2,3,4,5,6]
l1.index(3)
```

```python
# reverse -> permanently reverses list
l1=[1,2,3,4,5,6]
l1.reverse()
print(l1)
```

```python
# sort -> permanently sorts list
l1=[6,5,4,3,2,1]
l1.sort()
print(l1)
```

```python
# copy
l1=[1,2,3,4,5,6]
l2=l1.copy()
print(l2)
```

**LIST COMPREHENSION**
- a shortcut way of creating lists


```python
# scalar multiplication on a vector
v=[2,3,4]
s=-3
list1=[i*s for i in v] # i want this | from this | if condition (only if needed)
print(list1)
```

```python
# add squares
list1= [1,2,3,4,5]
list1=[i**2 for i in list1]
print(list1)
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
[num for num in range(1,51) if num%5==0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']
[language for language in languages if language.startswith('p' or 'P')]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]

# [fruit for fruit in my_fruits if fruit in basket and fruit.startswith('a')] also works but isnt nested if
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[(i*j)for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

** Traversing a list**
- itemwise
- indexwise

```python
# itemwise
L = [1,2,3,4]

for i in L:
    print(i)
```

```python
# indexwise
L = [1,2,3,4]

for i in range(0,len(L)):
    print(L[i])
```

**Zip function**
- combines items of lists together

```python
l1=[1,2,3,4]
l2=[-1,-2,-3,-4]

l3=list(zip(l1,l2))
print(l3)

# to add one l1[0] with l2[0] and l1[1] with l1[1] and so on, do below

l3=[(i+j) for i,j in l3]
print(l3)
```

```python
# imp note - since lists are mutable 

a=[1,2,3]
b=a

a.append(4)
print(a)
print(b) # changes are shown in both. so to not have changes shown in b, do below:
```

```python
a=[1,2,3]
b=a.copy()

a.append(4)
print(a)
print(b)
```
