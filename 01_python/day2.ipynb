3. **Type conversion and type casting** : changes one datatype to another ONLY if interchangeable
 - implicit (type conversion) : python automatically does the type conversion, No data loss
 - explicit (type casting): when we as a programmer do the necessary type conversion, May cause data loss, forced by programmer, creates new object, old object remains unchanged



```python
a=4.5
print(int(a)) # type casting done here. new integer created that can be used. original value of variable wont change.
print(a) # og value of a not changed
```

```python
# type conversion example
a = 5      # int
b = 2.5    # float

c = a + b  # int → float automatically
print(c)   # 7.5
```

```python
# type casting example
#Done manually by you
#Using built-in functions like:
#int()
#float()
#str()
d= 10.5
b= int(d) # You force the conversion using (int), and .75 gets lost
print(b)
print(d)
```
