**Operators**

```python
# Arithmetic operators
print(5-6)
print(5*6)
print(5/2)
print(5//2) # removes the decimal part, just gives integer part
print(5%2) # gives remainder
```

```python
# Relational operators | compares, returns true/false
print(9>10)
```

```python
# Logical operators | AND, OR, NOT | returns 1 or 0 or boolean
print(0 and 1)
print(0 or 1)
print(not 1)
```

```python
# bitwise operators | OR, AND, XOR (working done in notebook) | operate on binary values
print(2 & 3)
print(2 | 3)
print(2^3)
print(~3)
```

```python
# Assignment operators
a=2
a+=2
print(a)
# a++ or ++a like java ARE NOT USED IN PYTHON, creates ambiguity
```

```python
# Membership operators | in / not in
print("B" in "AbbA")
print("B" in "ABBA")
print(1 in [2,3,4,5])
```

```python
# Program - WAP to find the sum of a 3 digit number given by a user using operators | number entered is 345
number= int(input("enter a 3 digit number"))
a=number%10 #5
number= number//10 #34
b=number%10 #4
number=number//10 #3
c=number%10 #3
print(a+b+c)
```

**IF-ELSE | ELIF**

```python
# syntax for java:
# if (condition) {
     #code
  #} else{
       #code
  #}

# syntax for python:

if 1>2:
    print("true")
elif 1==2:
    print("equal")
else:
    print("false")

# instead of && used in java, here we use the word "and"
```

```python
# if else examples
# find the min of 3 given numbers
num1= int(input("enter the first number"))
num2= int(input("enter the second number"))
num3= int(input("enter the third number"))
if num1<num2 and num1<num3:
    print("num1 is min")
elif num2<num1 and num2<num3:
    print("num2 is min")
else:
    print("num3 is min")

```
